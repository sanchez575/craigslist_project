```python
#Q1-1,2
import random
import numpy

def iterate(times, maxNum):
    scores = []
    
    for n in range(times):
        
        s = 0
        
        while s < maxNum:
            r = random.randint(0,6)
            s += 2**r
        scores.append(s - maxNum)
    
    print("N: " + str(maxNum) + ", Mean: " + str(numpy.mean(scores)) + ", StDev: " + str(numpy.std(scores)))

print(iterate(1000,21))
print(iterate(100000,21))

import itertools
nums = [2**i for i in range(0,7)]


def run(n):
    s = 0
    for i in range(1,9):
        tups = list(itertools.product(nums,repeat=i))
        for tup in tups:
            if sum(tup) >= n and sum(tup[:-1]) < n:
                s += (sum(tup)-n)*(1/7.0)**(i)
        print(s)
    return s

run(21)



    
```

```python
#Q1-5,6
import random
import numpy

def conditionalNom(times, maxNum):
    scores = []
    newMax = maxNum - 8
    
    for n in range(times):
        
        s = 0
        
        while s < newMax:
            r = random.randint(0,6)
            s += 2**r
        if (s - newMax<= 5):
            scores.append(s - newMax)
            
    return len(scores) / float(times)

def probHas8(times, maxNum):
    has8 = 0
    
    for n in range(times):
        
        s = 0
        
        while s < maxNum:
            r = random.randint(0,6)
            s += 2**r
            if r == 3: has8 += 1
            
    return has8 / float(times)

print(conditionalNom(100000,21) / probHas8(100000,21))
print(conditionalNom(100000,1000) / probHas8(100000,1000))
```

```python
#Q1-7,8
eVal = sum([(2**i)*1.0/7 for i in range(7)])

def iterate2(times, maxNum):
    scores = []
    
    for n in range(times):
        
        s = 0
        
        while s < maxNum:
            r = random.randint(0,6)
            if r == 0:
                subtract = maxNum - (s + 11)
                d = 11
                if subtract > 0:
                    if subtract <= eVal:
                        d = 1
                else:
                    if abs(subtract) >= eVal:
                        d = 1
                s += d
                
            else:
                s += 2**r
        scores.append(s - maxNum)
    
    print("N: " + str(maxNum) + ", Mean: " + str(numpy.mean(scores)) + ", StDev: " + str(numpy.std(scores)))

iterate2(10000,21)
iterate2(100000,21)
```

```python
%%time
#Q3
#import data
import requests 
from bs4 import BeautifulSoup 
import re


def parsed_posts(soup_p, zip_code):
    all_posts = []

    for post in soup_p:
        d = {}
        d['title'] = post.find(class_='hdrlnk').contents[0]
        d['id'] = post.find(class_='hdrlnk')['data-id']
        d['zip'] = zip_code
        d['expected_ppftsq'] = zip_prices[zip_code]
        mp = post.find(class_='price')
        if mp: d['monthly_price'] = float(mp.contents[0][1:])

        s = str(post.find(class_='housing')) 
        m = re.search('(?<="housing">)(.*)(?=<sup>)', s)
        if m:
            s = m.group(1)
            br = re.search('[0-9]br', s)

            if br: d['bedrooms'] = float(br.group(0)[:-2])
            ft = re.search('[0-9]+ft', s)
            if ft: d['squared_feet'] = float(ft.group(0)[:-2])

        #n = post.find('small')
        #if n:
        #    d['neighborhood'] = n.contents[0][2:-1]
        im = post.find(class_='p').contents[0]
        if im.strip() != '':
            d['has_image'] = 1
        else:
            d['has_image'] = 0
        all_posts.append(d)
    
    return all_posts 

#all data
zip_codes = zip_prices.keys()
agg_posts = []

for z in zip_codes:
    
    current_post = 0
    
    baseURL = 'http://sfbay.craigslist.org/search/apa?is_paid=all&search_distance=0&search_distance_type=mi&postal=' + z
    URL = baseURL + '&s=0'

    response = requests.get(URL)

    html = response.content
    soup = BeautifulSoup(html)
    soup_posts = soup.findAll('p')

    while len(soup_posts) > 0:
        agg_posts.extend(parsed_posts(soup_posts, z))
        current_post += 100
        URL = baseURL + '&s=' + str(current_post)

        response = requests.get(URL)

        html = response.content
        soup = BeautifulSoup(html)
        soup_posts = soup.findAll('p')
```

```python
#define cosine string similarities
import re, math
from collections import Counter

WORD = re.compile(r'\w+')

def get_cosine(vec1, vec2):
     intersection = set(vec1.keys()) & set(vec2.keys())
     numerator = sum([vec1[x] * vec2[x] for x in intersection])

     sum1 = sum([vec1[x]**2 for x in vec1.keys()])
     sum2 = sum([vec2[x]**2 for x in vec2.keys()])
     denominator = math.sqrt(sum1) * math.sqrt(sum2)

     if not denominator:
        return 0.0
     else:
        return float(numerator) / denominator

def text_to_vector(text):
    words = WORD.findall(text)
    return Counter(words)

get_cosine(text_to_vector('San Francisco, Poor credit accepted '),text_to_vector('San Francisco'))
```

```python
#average neighborhood prices form rentjungle.com
neighborhoods = ['North Beach','Russian Hill','Castro-Upper Market','Noe Valley','Financial District',
                 'Mission','Presidio Heights','Bernal Heights','Haight-Ashbury','Potrero Hill','Pacific Heights',
                 'Inner Richmond','Twin Peaks','Glen Park','Marina','Western Addition','South Of Market',
                 'Chinatown','Inner Sunset','Nob Hill','Seacliff','Outer Richmond','West Of Twin Peaks',
                 'Parkside','Downtown','Excelsior','Outer Mission','Visitacion Valley','Bayview']
av_rent_prices = [4620,4571,4490,4453,4442,4440,4427,4396,4384,4315,4115,4102,4076,4045,4008,3905,3729,
                  3618,3561,3535,3493,3484,3452,3308,3254,3252,3234,2977,2686]

#deduplicate neighborhood names
def closest(x, possibilities):
    scores = [(0.0,x)]
    if x and not pd.isnull(x):
        for i,n in enumerate(possibilities):
            cosine = get_cosine(text_to_vector(x.lower()),text_to_vector(n.lower()))
            if cosine > 0.5:
                scores.append((cosine,possibilities[i]))
    return max(scores)[1]

cl['_neighborhood'] = cl['neighborhood'].apply(lambda x: closest(x,neighborhoods))
av_prices = pd.DataFrame({'_neighborhood':neighborhoods,'mean_rent':av_rent_prices})
df9 = pd.merge(cl,av_prices,how='left',on='_neighborhood')
```

```python
import numpy as np
#fill bedrooms and ft^2
beds_df = df[df.bedrooms.apply(lambda x: not pd.isnull(x))][['bedrooms','monthly_price']]
y = beds_df.bedrooms.values.reshape(-1, 1)
x = beds_df.monthly_price.values.reshape(-1, 1)

from sklearn import datasets, linear_model
regr_bed = linear_model.LinearRegression()
regr_bed.fit(x, y)

ft_df = df[df.squared_feet.apply(lambda x: not pd.isnull(x))][['squared_feet','monthly_price']]
y = ft_df.squared_feet.values.reshape(-1, 1)
x = ft_df.monthly_price.values.reshape(-1, 1)

regr_ft = linear_model.LinearRegression()
regr_ft.fit(x, y)

def apply_regr(row):
    if not pd.isnull(row['monthly_price']):
        if pd.isnull(row['bedrooms']):
            row['bedrooms'] = round(regr_bed.predict(np.array(row['monthly_price']).reshape(-1, 1))[0])
        if pd.isnull(row['squared_feet']):
            row['squared_feet'] = regr_ft.predict(np.array(row['monthly_price']).reshape(-1, 1))[0][0]
    return row
df = df.apply(apply_regr, axis=1)
```
